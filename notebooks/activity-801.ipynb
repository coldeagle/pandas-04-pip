```python
# Step 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 3
file_path = 'https://raw.githubusercontent.com/PacktWorkshops/The-pandas-Workshop/master/Chapter08/Data/PUF2020final_v1coll.csv'
df = pd.read_csv(file_path, delimiter=',')

# Step 4
df = df[["REGION", "SQFT", "BEDROOMS", "PRICE"]]

# Step 5
df.head(10)

# Step 6
df.PRICE.plot(kind='hist')

# Step 7
df.SQFT.plot(kind='hist')

# Step 8
df.plot(kind='scatter', x='SQFT', y='PRICE')

# Step 9
df.boxplot(by='BEDROOMS', column='PRICE')

# Step 10
df.plot(kind='scatter', x='BEDROOMS', y='PRICE')

# Step 11
df.boxplot(by='REGION', column='PRICE')

# Step 12
df.plot(kind='scatter', x='REGION', y='PRICE')

# Step 13
pd.pivot_table(df, index=['REGION']).PRICE.plot(kind='barh')

# Step 14
pd.pivot_table(df, index=['REGION']).PRICE.plot(kind='line')


```
